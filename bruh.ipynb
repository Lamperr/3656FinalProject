```julia
using LinearAlgebra

function GenerateRandUnitaries(n)
    [Matrix(qr(rand(ComplexF16, (2^i,2^i))).Q) for i=1:n]
end

UList = GenerateRandUnitaries(10)
display(UList)
[U'U ≈ I for U in UList]
```

```julia
function qr_householder(A)
    m, n = size(A)
    R = copy(A)
    V = [] # list of reflectors
    for j in 1:n
        # Build v for the current column
        v = copy(R[j:end, j])
        v[1] += sign(v[1]) * norm(v) # Here's the fix!
        v = normalize(v)
        # Build and apply F
        R[j:end, j:end] -= 2 * v * (v' * R[j:end,j:end])
        # Keep track of the reflectors
        push!(V, v)
    end
    V, R
end

```

```julia
function NewHouseholderQR(A)
    n,_ = size(A)
    
    for nb in 1:(n-1)  
        x = A[nb:n, nb] # extract the 1st column
        v = copy(x)
        
        v[1] += sign(x[1]) * norm(x) # get householder vector, first element is the norm of the whole vector
        v = v / norm(v) # normalize the vector
        
        A[nb:n, nb:n] -= 2 * v * (v' * A[nb:n, nb:n]) # apply the householder transformation to submatrix
        # * * *
        # 0 * * 
        # 0 0 *
        #       ...n
    end
    
    A
end
```
